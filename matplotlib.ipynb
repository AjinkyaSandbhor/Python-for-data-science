```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(4,15,100)
```

```python
y=np.sin(x)
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,label="Sin Wave", linestyle="-",linewidth=2, color="blue")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin Wave")
plt.legend()
plt.grid(True)
```

```python
x=np.random.rand(100)
y=np.random.rand(100)
color=np.random.rand(100)
size=1000*np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y,c=color,s=size, alpha=0.5,cmap="plasma")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("scatter plot")
plt.colorbar(label="color intensity")
plt.grid(True)
plt.show()
```

```python
angles =np.linspace(0,2*np.pi,100)
radii=np.random.rand(100)
```

```python
plt.figure(figsize=(5,4))
ax = plt.subplot(111, polar=True)
ax.plot(angles, radii, color="blue", linewidth=0.2)
ax.fill(angles, radii, color="blue", alpha=0.3)
plt.title("polar plot")
plt.show()
```
